```python
import matplotlib.pyplot as plt
import pandas as pd
import gzip
```

```python
# Plotting gender stuff
df = pd.read_csv(gzip.open('data_cleaned/cleanedBigGun_gender.csv.gz'))
df.head()
```

```python
df.columns
```

```python
data = pd.DataFrame(df.groupby('year').sum()['# of Male Victims'])
data['# of Female Victims'] = df.groupby('year').sum()['# of Female Victims']
data['# of Male Suspects'] = df.groupby('year').sum()['# of Male Suspects']
data['# of Female Suspects'] = df.groupby('year').sum()['# of Female Suspects']
data.reset_index(inplace = True)
data.rename(columns={"year": "Year"}, inplace = True)
data
```

```python
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap

output_file("bars.html")

data.Year = data.Year.astype(str)
data['# of Male Victims'] = data['# of Male Victims'].astype(str)
data['# of Female Victims'] = data['# of Female Victims'].astype(str)

source = ColumnDataSource(data)

p = figure(plot_width=800, plot_height=300, title="TEST",
           x_range=data.Year.tolist(), toolbar_location=None, tools="")

l = ['# of Male Victims', '# of Female Victims']

# p.vbar_stack(years, x='fruits', width=0.9, color=colors, source=source,
#              legend=[value(x) for x in years])

p.vbar_stack(l, x='Year', width=0.8, color=["black", "red"], source=source, legend=[value(x) for x in l])
# p.vbar(x='Year', top='# of Male Victims', width=.5, source=source,
#        line_color="white", fill_color='lightskyblue')

p.y_range.start = 0
p.x_range.range_padding = 0.05
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Year"
#p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None

# p.add_tools(HoverTool(tooltips=[("Year", "@Year"), ("# of Male Victims", "@{# of Male Victims}{f}")]))

show(p)
```
